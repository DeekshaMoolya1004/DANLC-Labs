```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame
sns.set_style('whitegrid')  # Optional: Set the style for the plots

# Function to plot bar chart for categorical columns
def plot_bar_chart(df, column_name, title=None, palette='Set2', rotation=45):
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=column_name, palette=palette)
    plt.title(title if title else f'Count of {column_name}', fontsize=14, weight='bold')
    plt.xlabel(column_name, fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=rotation)  # Rotate x-axis labels
    plt.tight_layout()  # Adjust layout for better visuals
    plt.show()

# Example usage
plot_bar_chart(df, 'Have you ever had suicidal thoughts ?',
               title='Distribution of Responses to Suicidal Thoughts', palette='coolwarm')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame
sns.set_style('whitegrid')  # Optional: Set the style for the plots

# Line chart for sleep schedule
def plot_sleep_schedule(df, time_column, sleep_column):
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df, x=time_column, y=sleep_column, marker="o", color="b", label="Sleep Duration")
    plt.title('Sleep Schedule Over Time', fontsize=14, weight='bold')
    plt.xlabel(time_column, fontsize=12)
    plt.ylabel('Sleep Duration (Hours)', fontsize=12)
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()  # Adjust layout for better visuals
    plt.show()

# Example usage (replace 'id' with a column representing time/order in your data)
plot_sleep_schedule(df, 'id', 'Sleep Duration')  # Replace 'id' with the appropriate time/order column
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to create a pie chart
def plot_pie_chart(df, column_name, title=None):
    # Get value counts for the column
    counts = df[column_name].value_counts()

    # Create the pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set2.colors)
    plt.title(title if title else f'Distribution of {column_name}', fontsize=14, weight='bold')
    plt.show()

# Example usage
plot_pie_chart(df, 'Have you ever had suicidal thoughts ?',
               title='Responses to Suicidal Thoughts Question')
```

```python
def plot_family_history_pie(df):
    counts = df['Family History of Mental Illness'].value_counts()
    plt.figure(figsize=(8, 8))
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)
    plt.title('Distribution of Family History of Mental Illness', fontsize=14, weight='bold')
    plt.show()

# Example usage
plot_family_history_pie(df)
```

```python
def plot_work_study_vs_satisfaction(df):
    plt.figure(figsize=(8, 6))
    sns.barplot(data=df, x='Study Satisfaction', y='Work/Study Hours', palette='viridis')
    plt.title('Average Work/Study Hours by Study Satisfaction', fontsize=14, weight='bold')
    plt.xlabel('Study Satisfaction', fontsize=12)
    plt.ylabel('Work/Study Hours', fontsize=12)
    plt.show()

# Example usage
plot_work_study_vs_satisfaction(df)
```

```python
def plot_work_study_line_chart(df):
    avg_hours_by_age = df.groupby('Age')['Work/Study Hours'].mean()
    plt.figure(figsize=(8, 6))
    plt.plot(avg_hours_by_age.index, avg_hours_by_age.values, marker='o', color='purple')
    plt.title('Average Work/Study Hours by Age', fontsize=14, weight='bold')
    plt.xlabel('Age', fontsize=12)
    plt.ylabel('Average Work/Study Hours', fontsize=12)
    plt.grid(alpha=0.3)
    plt.show()

# Example usage
plot_work_study_line_chart(df)
```
